```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PM2 = pd.read_csv('../Data/clean_Mean population exposure to PM2.5.csv')
```

```python
PM2.head()
```

```python
#PM2.set_index('Year', inplace=True)
PM2.groupby('Country')['Value'].plot(legend=True)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
#Median over all the countries
PM2.reset_index(inplace=True)
sns.lineplot(x=PM2['Year'], y=PM2['Value'], data=PM2)
plt.show() #increase :(
```

```python
sns.violinplot(x=PM2['Country'], y=PM2['Value'], data=PM2)
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
#USA
#PM2.set_index('Year', inplace=True)
PM2[PM2['Country'] == 'United States']['Value'].plot()
plt.ylim(0,20)
plt.title("PM2.5 exposure in USA") #good job
```

```python
#France
#PM2.set_index('Year', inplace=True)
PM2[PM2['Country'] == 'France']['Value'].plot()
plt.ylim(0,20)
plt.title("PM2.5 exposure in France") #good job
```
