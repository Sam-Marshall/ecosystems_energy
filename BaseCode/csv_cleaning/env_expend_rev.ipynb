```python
import pandas as pd
```

```python
env_expend_rev_file = "../Data/original_env_expend_rev.csv"
```

```python
env_expend_rev_csv = pd.read_csv(env_expend_rev_file)
env_expend_rev_df = pd.DataFrame(env_expend_rev_csv)
```

```python
env_expend_rev_df = env_expend_rev_df.loc[env_expend_rev_df['Measure'] == 'Millions USD, 2010 PPP prices'].reset_index()
```

```python
env_expend_rev_df = env_expend_rev_df.rename(columns={'Value' : 'Millions USD, 2010 PPP prices', 'COU' : 'Country Code'})
```

```python
env_expend_rev_df.columns.values
```

```python
env_expend_rev_df = env_expend_rev_df[['Country Code', 'Country', 'Tables', 'Sector', 'Industry', 'Expenditure', 'Year', 'Millions USD, 2010 PPP prices', 'Flag Codes', 'Flags']]
```

```python
env_expend_rev_df['Flags'] = env_expend_rev_df['Flags'].fillna('Actual value')
env_expend_rev_df['Flag Codes'] = env_expend_rev_df['Flag Codes'].fillna('A')

env_expend_rev_df['Flags'].value_counts()
```
