```python
import pandas as pd
```

```python
tax_file = "../Data/original_env_tax_revenue.csv"
```

```python
tax_csv = pd.read_csv(tax_file)
```

```python
tax_df = pd.DataFrame(tax_csv)

tax_df.head()
```

```python
USD2010PPP_tax_rev_df = tax_df.loc[(tax_df['VAR'] == 'BASE_PPP'), :].reset_index()
percent_tax_rev_df = tax_df.loc[(tax_df['VAR'] == 'BASE_REV'), :].reset_index()
```

```python
USD2010PPP_tax_rev_df = USD2010PPP_tax_rev_df[['index', 'COU', 'Country', 'Domain', 'Year', 'Value', 'Flag Codes', 'Flags']]
USD2010PPP_tax_rev_df.head()
```

```python
USD2010PPP_tax_rev_df['Flags'] = USD2010PPP_tax_rev_df['Flags'].fillna('Actual value')
USD2010PPP_tax_rev_df['Flag Codes'] = USD2010PPP_tax_rev_df['Flag Codes'].fillna('A')
```

```python
USD2010PPP_tax_rev_df = USD2010PPP_tax_rev_df.rename(columns={'COU' : 'Country Code', 'Value' : 'Tax revenue, millions USD 2010 PPP'})
```

```python
USD2010PPP_tax_rev_df.head()
```

```python
USD2010PPP_tax_rev_df.to_csv("../Data/clean_env_USD2010PPP_tax_revenue.csv")
```

```python
percent_tax_rev_df.head()
```

```python
percent_tax_rev_df = percent_tax_rev_df[['COU', 'Country', 'Domain', 'Year', 'Value', 'Flag Codes', 'Flags']]
percent_tax_rev_df['Flags'] = percent_tax_rev_df['Flags'].fillna('Actual value')
percent_tax_rev_df['Flag Codes'] = percent_tax_rev_df['Flag Codes'].fillna('A')

percent_tax_rev_df = percent_tax_rev_df.rename(columns={'COU' : 'Country Code', 'Value' : '% Total Tax Rev'})
```

```python
percent_tax_rev_df.head()
```

```python
percent_tax_rev_df.to_csv("../Data/clean_env_percent_tax_revenue.csv")
```
