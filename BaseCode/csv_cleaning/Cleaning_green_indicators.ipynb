```python
import pandas as pd
```

```python
indicators = pd.read_csv('../Data/Green_headline_indicators.csv')
```

```python
indicators.head(500)
```

```python
del indicators['COU']
del indicators['VAR']
del indicators['YEA']
del indicators['Unit Code']
del indicators['PowerCode Code']
del indicators['Reference Period Code']
del indicators['Reference Period']
del indicators['Flag Codes']
del indicators['Flags']
del indicators['PowerCode']
```

```python
indicators['Variable'].value_counts()
```

```python
#groupby indicators and save as csv
for i, g in indicators.groupby('Variable'):
    if i == "Production-based CO2 productivity, GDP per unit of energy-related CO2 emissions":
        g.to_csv('../Data/clean_Production_based_CO2_productivity.csv'.format(i), header=True, index_label=True)
    elif i == "Mean population exposure to PM2.5":
        g.to_csv('../Data/clean_{}.csv'.format(i), header=True, index_label=True)
    elif i == 'Gain of natural and semi-natural vegetated land, % since 1992':
        g.to_csv('../Data/clean_Gainofland.csv'.format(i), header=True, index_label=True)
    elif i == "Loss of natural and semi-natural vegetated land, % since 1992":
        g.to_csv('../Data/clean_Lossofland.csv'.format(i), header=True, index_label=True)
```
