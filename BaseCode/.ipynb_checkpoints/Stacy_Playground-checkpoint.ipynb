```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tax_per = pd.read_csv("../Data/clean_env_percent_tax_revenue.csv")
tax_per_df = pd.DataFrame(tax_per)
tax_per_df.head()
```

```python
allCountries = tax_per_df.Country.unique()
```

```python
waste_file = pd.read_csv("../Data/clean_generatedwaste.csv")
waste_generated_df = pd.DataFrame(waste_file)
waste_generated_df = waste_generated_df.rename(columns={'Value': 'Tonnes of Municipal Waste Generated'})
waste_generated_df = waste_generated_df[['Country', 'Year', 'Tonnes of Municipal Waste Generated']]
waste_generated_df.head()
```

```python
waste_tax_rev_df = tax_per_df.loc[tax_per_df['Domain'] == 'Waste management', :].reset_index()
waste_tax_rev_df = waste_tax_rev_df[['Country', 'Domain', 'Year', '% Total Tax Rev']]
waste_tax_rev_df.head()
```

```python
stringency_file = pd.read_csv("../Data/clean_all_stringency.csv")
stringency_df = pd.DataFrame(stringency_file)
stringency_df = stringency_df.loc[stringency_df['Variable'] == 'Environmental Policy Stringency', :]
stringency_df = stringency_df.rename(columns={'Value' : 'Environmental Policy Stringency'})
stringency_df = stringency_df[['Country', 'Year', 'Environmental Policy Stringency']]
stringency_df.head()
```

```python
cntry_waste_tax_list = waste_tax_rev_df['Country'].unique()
print(cntry_waste_tax_list)
```

```python
cntry_waste_gen_list = waste_generated_df['Country'].unique()
print(cntry_waste_gen_list)
```

```python
cntry_stringency_list = stringency_df['Country'].unique()
print(cntry_stringency_list)
```

```python
print(len(cntry_waste_tax_list))
print(len(cntry_waste_gen_list))
print(len(cntry_stringency_list))
```

```python
shared_waste_countries = list(set(cntry_waste_gen_list) & set(cntry_waste_tax_list) & set(cntry_stringency_list))
print(shared_waste_countries)
```

```python
shared_waste_tax_rev_df = waste_tax_rev_df.loc[waste_tax_rev_df['Country'].isin(shared_waste_countries)]
shared_waste_generated_df = waste_generated_df.loc[waste_generated_df['Country'].isin(shared_waste_countries)]
shared_stringency_df = stringency_df.loc[stringency_df['Country'].isin(shared_waste_countries)]

shared_waste_generated_df.head()
```

```python
waste_merged_df = pd.merge(shared_waste_tax_rev_df, shared_waste_generated_df, on=['Country', 'Year'], \
        how='inner')
```

```python
waste_merged_df = pd.merge(waste_merged_df, shared_stringency_df, on=['Country', 'Year'], \
        how='inner')
```

```python
waste_merged_df.head()
```

```python
waste_tax_ave_df = waste_merged_df.groupby(['Country'])['% Total Tax Rev'].agg(['mean'])
no_waste_tax_countries = []

for x, country in zip(range(len(waste_tax_ave_df)), waste_tax_ave_df.index):
    if waste_tax_ave_df.iloc[x]['mean'] == 0:
        no_waste_tax_countries.append(country)
```

```python
#print(len(waste_merged_df))
waste_merged_df = waste_merged_df.loc[~waste_merged_df['Country'].isin(no_waste_tax_countries)]
#print(len(waste_merged_df))
```

```python
country_names = waste_merged_df['Country'].unique()
country_list = []
waste_list = []
tax_waste_list = []
year_list = []
stringency_list = []

for country in country_names:
    df = waste_merged_df.loc[waste_merged_df['Country'] == country, :]
    years = list(df['Year'].values)
    waste = list(df['Tonnes of Municipal Waste Generated'].values)
    tax = list(df['% Total Tax Rev'].values)
    stringency = list(df['Environmental Policy Stringency'].values)
    
    country_list.append(country)
    waste_list.append(waste)
    tax_waste_list.append(tax)
    year_list.append(years)
    stringency_list.append(stringency)
```

```python
def make_patch_spines_invisible(ax):
    ax.set_frame_on(True)
    ax.patch.set_visible(False)
    for sp in ax.spines.values():
        sp.set_visible(False)
        
for i in range(len(country_list)):
    x_axis = np.arange(year_list[i][0], (year_list[i][len(year_list[i])-1]) + 1, 1)
    fig, ax1 = plt.subplots()
    fig.subplots_adjust(right=0.75)
    ax2 = ax1.twinx()
    ax3 = ax1.twinx()
    
    ax3.spines["right"].set_position(("axes", 1.2))
    ax3.spines["right"].set_visible(True)
    make_patch_spines_invisible(ax3)
    
    plt1 = ax1.plot(year_list[i], tax_waste_list[i], linestyle='--', marker='o', color='b')
    plt2 = ax2.plot(year_list[i], waste_list[i], linestyle='--', marker='o', color='g')
    plt3 = ax3.plot(year_list[i], stringency_list[i], linestyle='--', marker='o', color='m')
    
    plots = plt1+plt2+plt3
    labels = ['% Taxes For Waste','Tonnes of Waste', 'Stringency']
    ax1.legend(plots, labels, loc='upper center', bbox_to_anchor=(0.5, -0.22),
          fancybox=True, shadow=True, ncol=3)
    ax1.set_xticks(x_axis)
    ax1.set_xticklabels(x_axis, rotation=40)
    ax1.set_title(f"{country_list[i]} Tonnes of Waste and % Waste Taxes Per Year")
    ax1.set_xlabel('Years')
    ax1.set_ylabel('% Taxes for Waste Management', color='b')
    ax2.set_ylabel('Tonnes of Waste', color='g')
    ax3.set_ylabel('Stringency', color='m')
    ax1.tick_params(axis='y', colors='b')
    ax2.tick_params(axis='y', colors='g')
    ax3.tick_params(axis='y', colors='m')
```

```python
shared_ozone_countries = list(set(cntry_ozone_tax_list) & set(cntry_GHG_list) & set(cntry_stringency_list))
print(shared_ozone_countries)
```

```python
shared_ozone_tax_rev_df = ozone_tax_rev_df.loc[ozone_tax_rev_df['Country'].isin(shared_ozone_countries)]
shared_GHG_df = GHG_df.loc[GHG_df['Country'].isin(shared_ozone_countries)]
shared_ozone_stringency_df = stringency_df.loc[stringency_df['Country'].isin(shared_ozone_countries)]

shared_waste_generated_df.head()
```

```python
ozone_merged_df = pd.merge(shared_ozone_tax_rev_df, shared_GHG_df, on=['Country', 'Year'], \
        how='inner')
ozone_merged_df = pd.merge(ozone_merged_df, shared_ozone_stringency_df, on=['Country', 'Year'], \
        how='inner')
ozone_merged_df.head()
```

```python
ozone_tax_ave_df = ozone_merged_df.groupby(['Country'])['% Total Tax Rev'].agg(['mean'])
no_ozone_tax_countries = []

for x, country in zip(range(len(ozone_tax_ave_df)), ozone_tax_ave_df.index):
    if ozone_tax_ave_df.iloc[x]['mean'] == 0:
        no_ozone_tax_countries.append(country)

ozone_merged_df = ozone_merged_df.loc[~ozone_merged_df['Country'].isin(no_ozone_tax_countries)]
```

```python
for i in range(len(o_country_list)):
    x_axis = np.arange(o_year_list[i][0], (o_year_list[i][len(o_year_list[i])-1]) + 1, 1)
    fig, ax1 = plt.subplots()
    fig.subplots_adjust(right=0.75)
    ax2 = ax1.twinx()
    ax3 = ax1.twinx()
    
    ax3.spines["right"].set_position(("axes", 1.2))
    ax3.spines["right"].set_visible(True)
    make_patch_spines_invisible(ax3)
    
    plt1 = ax1.plot(o_year_list[i], tax_ozone_list[i], linestyle='--', marker='o', color='b')
    plt2 = ax2.plot(o_year_list[i], ozone_list[i], linestyle='--', marker='o', color='g')
    plt3 = ax3.plot(o_year_list[i], o_stringency_list[i], linestyle='--', marker='o', color='m')
    
    plots = plt1+plt2+plt3
    labels = ['% Taxes For Ozone','Tonnes of CO2 equivalent', 'Stringency']
    ax1.legend(plots, labels, loc='upper center', bbox_to_anchor=(0.5, -0.22),
          fancybox=True, shadow=True, ncol=3)
    ax1.set_xticks(x_axis)
    ax1.set_xticklabels(x_axis, rotation=40)
    ax1.set_title(f"{o_country_list[i]} Tonnes of CO2 equivalent and % Ozone Taxes Per Year")
    ax1.set_xlabel('Years')
    ax1.set_ylabel('% Taxes for Ozone', color='b')
    ax2.set_ylabel('Tonnes of CO2 equivalent', color='g')
    ax3.set_ylabel('Stringency', color='m')
    ax1.tick_params(axis='y', colors='b')
    ax2.tick_params(axis='y', colors='g')
    ax3.tick_params(axis='y', colors='m')
```
