```python
del emission['COU']
del emission['POL']
del emission['Pollutant']
del emission['VAR']
del emission['YEA']
del emission['Unit Code']
del emission['PowerCode Code']
del emission['Reference Period Code']
del emission['Reference Period']
del emission['Flag Codes']
del emission['Flags']
del emission['PowerCode']
```

```python
#the total emission is the only one we want?

for i, g in emission.groupby('Variable'):
    if i == "Total  emissions excluding LULUCF":   
        g.to_csv('../Data/clean_TotalGazEmission.csv'.format(i), header=True, index_label=True)
```
