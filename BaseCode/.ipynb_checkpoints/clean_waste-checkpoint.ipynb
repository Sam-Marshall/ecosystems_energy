```python
import pandas as pd
```

```python
recycling = pd.read_csv('../Data/Waste_municipal_treatment.csv')
```

```python
recycling.head()
```

```python
del recycling['COU']
del recycling['VAR']
del recycling['YEA']
del recycling['Unit Code']
del recycling['PowerCode Code']
del recycling['Reference Period Code']
del recycling['Reference Period']
del recycling['Flag Codes']
del recycling['Flags']
del recycling['PowerCode']
```

```python
recycling['Variable'].value_counts()
```

```python
#Municipal waste generated per capita and Municipal waste treated  is the only one we want?

for i, g in recycling.groupby('Variable'):
    if i == "Municipal waste generated":   
        g.to_csv('../Data/clean_generatedwaste.csv', header=True, index_label=True)
    if i == "Municipal waste treated":
        g.to_csv('../Data/clean_recycledwaste.csv', header=True, index_label=True)
```

```python
#percentage waste

generated = pd.read_csv('../Data/clean_generatedwaste.csv')
recycled = pd.read_csv('../Data/clean_recycledwaste.csv')

generated['recycled'] = recycled['Value']

generated['percentofrecycledwaste'] = round(generated['recycled']/generated['Value']*100,2)

generated.to_csv('../Data/clean_percentofrecycledwaste.csv',header=True)
```
