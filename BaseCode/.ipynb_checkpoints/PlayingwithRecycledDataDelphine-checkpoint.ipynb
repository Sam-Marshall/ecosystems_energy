```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
recycling = pd.read_csv('../Data/clean_percentofrecycledwaste.csv')
```

```python
recycling.head(20)
```

```python
#filter crazy values
recycling = recycling[recycling["percentofrecycledwaste"] < 120]
recycling.head()
```

```python
recycling.set_index('Year', inplace=True)
recycling.groupby('Country')['percentofrecycledwaste'].plot(legend=True)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
#Median over all the countries
recycling.reset_index(inplace=True)
sns.lineplot(x=recycling['Year'], y=recycling['percentofrecycledwaste'], data=recycling)
plt.show()
```

```python
sns.violinplot(x=recycling['Country'], y=recycling['percentofrecycledwaste'], data=recycling)
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
#by country
recycling[recycling['Country'] == 'Austria']['percentofrecycledwaste'].plot()
plt.ylim(0,120)
plt.title("Percent of recycled waste in Austria")
```

```python
#USA, clearly some data point are missing
recycling[recycling['Country'] == 'United States']['percentofrecycledwaste'].plot()
plt.ylim(0,120)
plt.title("Percent of recycled waste in USA")
```
