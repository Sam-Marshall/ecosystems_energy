```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from functools import reduce
```

```python
#import biodiversity file

biodiversity = pd.read_csv('LPI_biodiversity.csv')
biodiversity.head()
```

```python
biodiversity.columns
```

```python
del biodiversity['Binomial']
del biodiversity['Reference']
del biodiversity['Class']
del biodiversity['Order']
del biodiversity['Family']
del biodiversity['Genus']
del biodiversity['Subspecies']
del biodiversity['Authority']
del biodiversity['Location']
del biodiversity['All_countries']
del biodiversity['Specific_location']
del biodiversity['T_realm']
del biodiversity['T_biome']
del biodiversity['FW_realm']
del biodiversity['FW_biome']
del biodiversity['M_realm']
del biodiversity['M_ocean']
del biodiversity['M_biome']
del biodiversity['Method']
```

```python
biodiversity.head()
```

```python
#Cleaning to do in the Units column
biodiversity['Units'].value_counts()

#The most consistent value are a count of individuals, so I am going to remove all other units (example:pairs, nest...)
#I am keeping only the line with individuals value count 

biodiversity_clean = biodiversity[(biodiversity.Units == 'Number of individuals')|(biodiversity.Units == 'Individuals')|\
                                  (biodiversity.Units == 'number of individuals')|(biodiversity.Units == 'Population estimate')|\
                                  (biodiversity.Units == 'Number of observed birds')|(biodiversity.Units == 'Mean abundance')|\
                                  (biodiversity.Units == 'Total count')|(biodiversity.Units == 'number of fish')|\
                                  (biodiversity.Units == 'Total number of individuals')|(biodiversity.Units == 'Number of birds')\
                                 |(biodiversity.Units == 'Total number of birds')|(biodiversity.Units == 'Estimated number of individuals')\
                                 |(biodiversity.Units == 'Individual fish')|(biodiversity.Units == 'Total number of individuals captured')\
                                 |(biodiversity.Units == 'Average number of individuals')|(biodiversity.Units == 'Estimated abundance')\
                                 |(biodiversity.Units == 'Individual')|(biodiversity.Units == 'Population size')\
                                 |(biodiversity.Units == 'Total numbers')|(biodiversity.Units == 'Number of individual frogs observed')\
                                 |(biodiversity.Units == 'Annual average number of individuals')|(biodiversity.Units == 'Individual birds')\
                                 |(biodiversity.Units == 'Mean number of individuals')|(biodiversity.Units == 'mean abundance')\
                                 |(biodiversity.Units == 'individuals')|(biodiversity.Units == 'Abundance')\
                                 |(biodiversity.Units == 'Number of independent tigers')|(biodiversity.Units == 'Tiger abundance')\
                                 |(biodiversity.Units == 'Number of individuals observed')|(biodiversity.Units == 'Individual fish')\
                                 |(biodiversity.Units == 'Individual counts')|(biodiversity.Units == 'Avr. Number Individuals')\
                                 |(biodiversity.Units == 'Estimate - entire population')|(biodiversity.Units == 'Estimated population size')\
                                 |(biodiversity.Units == 'number of frogs')|(biodiversity.Units == 'Mean abundance of individuals')\
                                 |(biodiversity.Units == 'population estimate')] 
                        

#test if it worked
biodiversity_clean['Units'].count()
```

```python
biodiversity_clean.head()
```

```python
#Create a df for each year summing the number of species counted
biodiversity_bycountry_1990 = biodiversity_clean.groupby(['Country','System'],as_index=False)['1990'].sum()
biodiversity_bycountry_1991 = biodiversity_clean.groupby(['Country'],as_index=False)['1991'].sum()
biodiversity_bycountry_1992 = biodiversity_clean.groupby(['Country'],as_index=False)['1992'].sum()
biodiversity_bycountry_1993 = biodiversity_clean.groupby(['Country'],as_index=False)['1993'].sum()
biodiversity_bycountry_1994 = biodiversity_clean.groupby(['Country'],as_index=False)['1994'].sum()
biodiversity_bycountry_1995 = biodiversity_clean.groupby(['Country'],as_index=False)['1995'].sum()
biodiversity_bycountry_1996 = biodiversity_clean.groupby(['Country'],as_index=False)['1996'].sum()
biodiversity_bycountry_1997 = biodiversity_clean.groupby(['Country'],as_index=False)['1997'].sum()
biodiversity_bycountry_1998 = biodiversity_clean.groupby(['Country'],as_index=False)['1998'].sum()
biodiversity_bycountry_1999 = biodiversity_clean.groupby(['Country'],as_index=False)['1999'].sum()
biodiversity_bycountry_2000 = biodiversity_clean.groupby(['Country'],as_index=False)['2000'].sum()
biodiversity_bycountry_2001 = biodiversity_clean.groupby(['Country'],as_index=False)['2001'].sum()
biodiversity_bycountry_2002 = biodiversity_clean.groupby(['Country'],as_index=False)['2002'].sum()
biodiversity_bycountry_2003 = biodiversity_clean.groupby(['Country'],as_index=False)['2003'].sum()
biodiversity_bycountry_2004 = biodiversity_clean.groupby(['Country'],as_index=False)['2004'].sum()
biodiversity_bycountry_2005 = biodiversity_clean.groupby(['Country'],as_index=False)['2005'].sum()
biodiversity_bycountry_2006 = biodiversity_clean.groupby(['Country'],as_index=False)['2006'].sum()
biodiversity_bycountry_2007 = biodiversity_clean.groupby(['Country'],as_index=False)['2007'].sum()
biodiversity_bycountry_2008 = biodiversity_clean.groupby(['Country'],as_index=False)['2008'].sum()
biodiversity_bycountry_2009 = biodiversity_clean.groupby(['Country'],as_index=False)['2009'].sum()
biodiversity_bycountry_2010 = biodiversity_clean.groupby(['Country'],as_index=False)['2010'].sum()
biodiversity_bycountry_2011 = biodiversity_clean.groupby(['Country'],as_index=False)['2011'].sum()
biodiversity_bycountry_2012 = biodiversity_clean.groupby(['Country'],as_index=False)['2012'].sum()
biodiversity_bycountry_2013 = biodiversity_clean.groupby(['Country'],as_index=False)['2013'].sum()

df_list = [biodiversity_bycountry_1990,biodiversity_bycountry_1991,biodiversity_bycountry_1992,biodiversity_bycountry_1993,\
          biodiversity_bycountry_1994,biodiversity_bycountry_1995,biodiversity_bycountry_1996,biodiversity_bycountry_1997,\
          biodiversity_bycountry_1998,biodiversity_bycountry_1999,biodiversity_bycountry_2000,biodiversity_bycountry_2001,\
          biodiversity_bycountry_2002,biodiversity_bycountry_2003,biodiversity_bycountry_2004,biodiversity_bycountry_2005,\
          biodiversity_bycountry_2006,biodiversity_bycountry_2007,biodiversity_bycountry_2008,biodiversity_bycountry_2009,\
          biodiversity_bycountry_2010,biodiversity_bycountry_2011,biodiversity_bycountry_2012,biodiversity_bycountry_2013]

biodiversity_bycountry_all = reduce(lambda  left,right: pd.merge(left,right,on=['Country'],
                                            how='inner'), df_list)

biodiversity_bycountry_all = biodiversity_bycountry_all.set_index(['Country','System'])

biodiversity_bycountry_all = biodiversity_bycountry_all.replace(0.000000e+00, np.nan)

biodiversity_bycountry_all
```

```python
#plot Argentina

Argentina_Marine = pd.DataFrame(biodiversity_bycountry_all.loc["Argentina","Marine"]).reset_index()

Argentina_Marine['Years'] = Argentina_Marine.iloc[:,0]
Argentina_Marine['ArgentinaMarine'] = Argentina_Marine.iloc[:,1]

plt.plot(Argentina_Marine['Years'],Argentina_Marine['ArgentinaMarine'])

#plot Austria

Austria_Freshwater = pd.DataFrame(biodiversity_bycountry_all.loc["Austria","Freshwater"]).reset_index()

Austria_Freshwater['Years'] = Austria_Freshwater.iloc[:,0]
Austria_Freshwater['AustriaFreshwater'] = Austria_Freshwater.iloc[:,1]

plt.plot(Austria_Freshwater['Years'],Austria_Freshwater['AustriaFreshwater'])

#plot France
France_Marine = pd.DataFrame(biodiversity_bycountry_all.loc["France","Marine"]).reset_index()

France_Marine['Years'] = France_Marine.iloc[:,0]
France_Marine['FranceMarine'] = France_Marine.iloc[:,1]

plt.plot(France_Marine['Years'],France_Marine['FranceMarine'])

#plot Finland
Finland_Freshwater = pd.DataFrame(biodiversity_bycountry_all.loc["Finland","Freshwater"]).reset_index() #good !
Finland_Freshwater['Years'] = France_Marine.iloc[:,0]
Finland_Freshwater['FinlandFreshwater'] = Finland_Freshwater.iloc[:,1]

plt.plot(Finland_Freshwater['Years'],Finland_Freshwater['FinlandFreshwater'])

plt.legend()
plt.tick_params(axis='x', rotation=45)
```
