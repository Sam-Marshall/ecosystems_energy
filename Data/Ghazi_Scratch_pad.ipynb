```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
bio = pd.read_csv('../Data/clean_biodiversity.csv')
bio1 = pd.read_csv('../Data/clean_biodiversity_nogroupby.csv')
water = pd.read_csv('../Data/Total_FRESHWATER/all_freshwater_abstract.csv')
```

```python
#Collecting Fish in Freshwater
freshwater_fish0 = bio1.loc[bio1["System"]=="Freshwater",:]
freshwater_fish1 = freshwater_fish0.loc[:,"2008":"2015"]
freshwater_fish2 = freshwater_fish0[['Country']]
freshwater_fish3 = freshwater_fish0.loc[:,"1990":"2015"]

#bioheaders = bio1[['Country','System']]
#bio1
freshwater_fish = freshwater_fish2.join(freshwater_fish1)
freshwater_fish_years = freshwater_fish2.join(freshwater_fish3)

#Australia_fish = freshwater_fish0.loc[freshwater_fish0["Country"]=="Australia",:]
#Australia_fish
```

```python
c = np.arange(2008,2015,1)
c = c.tolist()
c = map(str,c)

d = np.arange(1990,2015,1)
d = d.tolist()
d = map(str,d)
```

```python
freshwater_fish = pd.melt(freshwater_fish, id_vars='Country', value_vars=c).round()
freshwater_fish_years = pd.melt(freshwater_fish_years, id_vars='Country', value_vars=d).round()
```

```python
#Drop Countries without Freshwater Fish Values
freshwater_fish.dropna(inplace=True)
freshwater_fish.to_csv('../Stuff/Merged_Bio_Water/freshwater_fish.csv', header=True, index_label=True)

freshwater_fish_years.dropna(inplace=True)
freshwater_fish_years.to_csv('../Stuff/Merged_Bio_Water/freshwater_fish_years.csv', header=True, index_label=True)
```

```python
country_we_want = ['Austria', 'United Kingdom', 'Canada', 'United States', 'Norway', 'Denmark', 'France', 'Australia','China',"China(People's Republic of)"]
idx = (freshwater_fish["Country"] == "United Kingdom")
for country in country_we_want:
    idx |= (freshwater_fish["Country"] == country)
freshwater_fish.reset_index()
```

```python
ids = (freshwater_fish_years["Country"] == "United Kingdom")
for country in country_we_want:
    ids |= (freshwater_fish_years["Country"] == country)
freshwater_fish_years.reset_index()
```

```python
#Pull out Canadian, USA, GREECE, AUSTRILIA, SPAIN Fishes
# Years 2008-2012 Only
Canadian_fish = freshwater_fish.loc[freshwater_fish["Country"]=="Canada",:].reset_index()
Canadian_fish = Canadian_fish.loc[18:22].reset_index()
Canadian_fish = Canadian_fish.rename(columns={'variable' : 'Year', 'value': 'Population'})
Canadian_fish = Canadian_fish[['Country', 'Year', 'Population']]
Canadian_fish.to_csv('../Stuff/Merged_Bio_Water/Canadian_fish.csv', header=True, index_label=True)


USA_fish = freshwater_fish.loc[freshwater_fish["Country"]=="United States",:].reset_index()
USA_fish = USA_fish.loc[USA_fish['value']>0].reset_index()
USA_fish = USA_fish.rename(columns={'variable' : 'Year', 'value': 'Population'})
USA_fish = USA_fish[['Country', 'Year', 'Population']]
USA_fish.to_csv('../Stuff/Merged_Bio_Water/USA_fish.csv', header=True, index_label=True)


Spain_fish = freshwater_fish.loc[freshwater_fish["Country"]=="Spain",:].reset_index()
Spain_fish = Spain_fish.loc[18:22].reset_index()
Spain_fish = Spain_fish.rename(columns={'variable' : 'Year', 'value': 'Population'})
Spain_fish = Spain_fish[['Country', 'Year', 'Population']]
Spain_fish.to_csv('../Stuff/Merged_Bio_Water/Spain_fish.csv', header=True, index_label=True)


Australia_fish = freshwater_fish.loc[freshwater_fish["Country"]=="Australia",:].reset_index()
#Australia_fish = Australia_fish.loc[18:22].reset_index()
Australia_fish = Australia_fish.rename(columns={'variable' : 'Year', 'value': 'Population'})
Australia_fish = Australia_fish[['Country', 'Year', 'Population']]
Australia_fish.to_csv('../Stuff/Merged_Bio_Water/Australia_fish.csv', header=True, index_label=True)


Greece_fish = freshwater_fish.loc[freshwater_fish["Country"]=="Greece",:].reset_index()
Greece_fish = Greece_fish.loc[18:22].reset_index()
Greece_fish = Greece_fish.rename(columns={'variable' : 'Year', 'value': 'Population'})
Greece_fish = Greece_fish[['Country', 'Year', 'Population']]
Greece_fish.to_csv('../Stuff/Merged_Bio_Water/Greece_fish.csv', header=True, index_label=True)

fish_they_care = freshwater_fish.loc[idx].reset_index()
fish_they_care = fish_they_care.rename(columns={'variable' : 'Year', 'value': 'Population'})
fish_they_care = fish_they_care[['Country', 'Year', 'Population']]
fish_they_care.to_csv('../Stuff/Merged_Bio_Water/fish_they_care.csv', header=True, index_label=True)

#Year 1990-2015
fish_they_care_year = freshwater_fish_years.loc[ids].reset_index()
fish_they_care_year = fish_they_care_year.rename(columns={'variable' : 'Year', 'value': 'Population'})
fish_they_care_year = fish_they_care_year[['Country', 'Year', 'Population']]
fish_they_care_year.to_csv('../Stuff/Merged_Bio_Water/fish_they_care_year.csv', header=True, index_label=True)



freshwater_fish_years = freshwater_fish_years.rename(columns={'variable' : 'Year', 'value': 'Population'})
freshwater_fish_years = freshwater_fish_years[['Country', 'Year', 'Population']]

freshwater_fish = freshwater_fish.rename(columns={'variable' : 'Year', 'value': 'Population'})
freshwater_fish = freshwater_fish[['Country', 'Year', 'Population']]
```
