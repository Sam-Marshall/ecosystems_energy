```python
import pandas as pd
file = "FRESHWATER_PULL.csv"
df_original = pd.read_csv(file)
```

```python
water = df_original[["Country","Source","Variable","Year","Unit","Value"]]
```

```python
water.head()
```

```python
# TOTAL Freshwater = LAND + GROUND

freshwater_all = water.loc[water["Source"] == "Total freshwater",:]
freshwater_all = freshwater_all.loc[freshwater_all["Value"] > 0,:]
freshwater_all = freshwater_all.reset_index(drop=True)

GROUP_freshwater_all=freshwater_all.groupby(["Year","Country","Variable"]).mean()

freshwater_total_abstract = freshwater_all.loc[freshwater_all["Variable"] == "Total gross abstraction",:]
freshwater_total_abstract = freshwater_all.loc[freshwater_all["Value"] > 0,:]
freshwater_total_abstract = freshwater_total_abstract.reset_index(drop=True)

freshwater_public = freshwater_all.loc[freshwater_all["Variable"] == "Public water supply",:]
freshwater_public = freshwater_all.loc[freshwater_all["Value"] > 0,:]
freshwater_public = freshwater_public.reset_index(drop=True)

freshwater_Agriculture = freshwater_all.loc[freshwater_all["Variable"] == "Agriculture, forestry, fishing",:]
freshwater_Agriculture = freshwater_all.loc[freshwater_all["Value"] > 0,:]
freshwater_Agriculture = freshwater_Agriculture.reset_index(drop=True)

freshwater_Industry = freshwater_all.loc[freshwater_all["Variable"] == "Manufacturing industry",:]
freshwater_Industry = freshwater_all.loc[freshwater_all["Value"] > 0,:]
freshwater_Industry = freshwater_Industry.reset_index(drop=True)

freshwater_irrigation = freshwater_all.loc[freshwater_all["Variable"] == "Irrigation",:]
freshwater_irrigation = freshwater_all.loc[freshwater_all["Value"] > 0,:]
freshwater_irrigation = freshwater_irrigation.reset_index(drop=True)

freshwater_abstraction_captia = freshwater_all.loc[freshwater_all["Variable"] == "Gross abstractions per capita",:]
freshwater_abstraction_captia = freshwater_all.loc[freshwater_all["Value"] > 0,:]
freshwater_abstraction_captia = freshwater_abstraction_captia.reset_index(drop=True)

freshwater_AQUAculture = freshwater_all.loc[freshwater_all["Variable"] == "Aquaculture",:]
freshwater_AQUAculture = freshwater_all.loc[freshwater_all["Value"] > 0,:]
freshwater_AQUAculture = freshwater_AQUAculture.reset_index(drop=True)

#Export to CSV
freshwater_all.to_csv('Total FRESHWATER/all_freshwater.csv', header=True, index_label=True)
GROUP_freshwater_all.to_csv('Total FRESHWATER/group_freshwater.csv', header=True, index_label=True)
freshwater_total_abstract.to_csv('Total FRESHWATER/all_freshwater_abstract.csv', header=True, index_label=True)
freshwater_public.to_csv('Total FRESHWATER/all_freshwater_public.csv', header=True, index_label=True)
freshwater_Agriculture.to_csv('Total FRESHWATER/all_freshwater_agriculture.csv', header=True, index_label=True)
freshwater_Industry.to_csv('Total FRESHWATER/all_freshwater_industry.csv', header=True, index_label=True)
freshwater_irrigation.to_csv('Total FRESHWATER/all_freshwater_irrigation.csv', header=True, index_label=True)
freshwater_abstraction_captia.to_csv('Total FRESHWATER/all_freshwater_capita.csv', header=True, index_label=True)
freshwater_AQUAculture.to_csv('Total FRESHWATER/all_freshwater_aquaculture.csv', header=True, index_label=True)

freshwater_total_abstract
```
