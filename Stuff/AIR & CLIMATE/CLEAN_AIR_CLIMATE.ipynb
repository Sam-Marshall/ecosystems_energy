```python
import pandas as pd
file = "GHG.csv"
df_original = pd.read_csv(file)
```

```python
df_original.head()
```

```python
pollutes = df_original[["Country","POL","Variable","Year","Unit","Value"]]
```

```python
#Pollutes = GHG
GHG = pollutes.loc[pollutes["POL"] == "GHG",:]
GHG = GHG.reset_index(drop=True)

#print(len(GHG))

GHG_including_LULUCF = GHG.loc[GHG["Variable"] == "Total  emissions including LULUCF",:]
GHG_including_LULUCF = GHG_including_LULUCF.reset_index(drop=True)

GHG_excluding_LULUCF = GHG.loc[GHG["Variable"] == "Total  emissions excluding LULUCF",:]
GHG_excluding_LULUCF = GHG_excluding_LULUCF.reset_index(drop=True)

GHG_2000 = GHG.loc[GHG["Variable"] == "Total GHG, Index 2000=100",:]
GHG_2000 = GHG_2000.reset_index(drop=True)

GHG_total_capita = GHG.loc[GHG["Variable"] == "Total GHG per capita",:]
GHG_total_capita = GHG_total_capita.reset_index(drop=True)

GHG_Energy = GHG.loc[GHG["Variable"]=="1 - Energy",:]
GHG_Energy = GHG_Energy.reset_index(drop=True)

GHG_Energy_MC = GHG.loc[GHG["Variable"]=="1A2 - Manufacturing industries and construction",:]
GHG_Energy_MC = GHG_Energy_MC.reset_index(drop=True)

GHG_Energy_Trans = GHG.loc[GHG["Variable"]=="1A3 - Transport",:]
GHG_Energy_Trans = GHG_Energy_Trans.reset_index(drop=True)

GHG_industrial = GHG.loc[GHG["Variable"]=="2- Industrial processes and product use",:]
GHG_industrial = GHG_industrial.reset_index(drop=True)

GHG_Agriculture = GHG.loc[GHG["Variable"]=="3 - Agriculture",:]
GHG_Agriculture = GHG_Agriculture.reset_index(drop=True)

GHG_Waste = GHG.loc[GHG["Variable"]=="5 - Waste",:]
GHG_Waste = GHG_Waste.reset_index(drop=True)

GHG_Other = GHG.loc[GHG["Variable"]=="6 - Other",:]
GHG_Other = GHG_Other.reset_index(drop=True)

#Export to CSV
GHG_including_LULUCF.to_csv('GHG_CLEAN_DATA/clean_GHG_including_LULUCF.csv', header=True, index_label=True)
GHG_excluding_LULUCF.to_csv('GHG_CLEAN_DATA/clean_GHG_excluding_LULUCF.csv', header=True, index_label=True)
GHG_2000.to_csv('GHG_CLEAN_DATA/clean_GHG_index_2000.csv', header=True, index_label=True)
GHG_Energy.to_csv('GHG_CLEAN_DATA/clean_GHG_Energy.csv', header=True, index_label=True)
GHG_Energy_Trans.to_csv('GHG_CLEAN_DATA/clean_GHG_Transport.csv', header=True, index_label=True)
GHG_industrial.to_csv('GHG_CLEAN_DATA/clean_GHG_industrial.csv', header=True, index_label=True)
GHG_Agriculture.to_csv('GHG_CLEAN_DATA/clean_GHG_Agriculture.csv', header=True, index_label=True)
GHG_Waste.to_csv('GHG_CLEAN_DATA/clean_GHG_Waste.csv', header=True, index_label=True)
GHG_Other.to_csv('GHG_CLEAN_DATA/clean_GHG_OtherEnergy.csv', header=True, index_label=True)
GHG_Energy_MC.to_csv('GHG_CLEAN_DATA/clean_GHG_Manufacturing_Construction.csv', header=True, index_label=True)
GHG.to_csv('GHG_CLEAN_DATA/clean_GHG.csv', header=True, index_label=True)
GHG_total_capita.to_csv('GHG_CLEAN_DATA/clean_GHG_per_capita.csv', header=True, index_label=True)
```

```python
#Pollutes = CO2
CO2 = pollutes.loc[pollutes["POL"] == "CO2",:]
CO2 = CO2.reset_index(drop=True)

CO2_excluding_LULUCF = CO2.loc[CO2["Variable"] == "Total  emissions excluding LULUCF",:]
CO2_excluding_LULUCF = CO2_excluding_LULUCF.reset_index(drop=True)

CO2_2000 = CO2.loc[CO2["Variable"] == "Total GHG, Index 2000=100",:]
CO2_2000 = CO2_2000.reset_index(drop=True)

#Export to CSV
CO2.to_csv('CO2_CLEAN_DATA/clean_CO2.csv', header=True, index_label=True)
CO2_excluding_LULUCF.to_csv('CO2_CLEAN_DATA/clean_CO2_exclude_LULUCF.csv', header=True, index_label=True)
CO2_2000.to_csv('CO2_CLEAN_DATA/clean_CO2_index_2000.csv', header=True, index_label=True)
```

```python
#Pollutes = Methane
CH4 = pollutes.loc[pollutes["POL"] == "CH4",:]
CH4 = CH4.reset_index(drop=True)

CH4_excluding_LULUCF = CH4.loc[CH4["Variable"] == "Total  emissions excluding LULUCF",:]
CH4_excluding_LULUCF = CH4_excluding_LULUCF.reset_index(drop=True)

CH4_2000 = CH4.loc[CH4["Variable"] == "Total GHG, Index 2000=100",:]
CH4_2000 = CH4_2000.reset_index(drop=True)

#Export to CSV
CH4.to_csv('CH4_CLEAN_DATA/clean_CH4.csv', header=True, index_label=True)
CH4_excluding_LULUCF.to_csv('CH4_CLEAN_DATA/clean_CH4_exclude_LULUCF.csv', header=True, index_label=True)
CH4_2000.to_csv('CH4_CLEAN_DATA/clean_CH4_index_2000.csv', header=True, index_label=True)
```
